```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from datasets import load_dataset, DatasetDict
from datasets import Dataset

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = load_dataset("Khalida1w/funny_quotes")
```

```python
dataset
```

```python
lengths_dataset_elements = [len(example['quote']) for example in dataset['train']]

# Créer un histogramme et une courbe de densité
plt.figure(figsize=(10,5))
sns.histplot(lengths_dataset_elements, kde=True, label='Train', color='blue')

plt.title("Density curve of Sentences length")
plt.xlabel("Sentences length")
plt.ylabel("Density")
plt.xlim(0, 1000)
plt.legend()
plt.show()
```

```python
max_length = 200

train_filtered = dataset.filter(lambda example: len(example['quote']) <= max_length)
```

```python
train_filtered["train"]
```

```python
splits_train_temp = train_filtered['train'].train_test_split(test_size=0.25, seed=42)

splits_val_test = splits_train_temp['test'].train_test_split(test_size=0.4, seed=42)

dataset_dict = DatasetDict({
    'train': splits_train_temp['train'],
    'val': splits_val_test['train'],
    'test': splits_val_test['test']
})
```

```python
dataset_dict
```

```python
# Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

tokenizer.pad_token = tokenizer.eos_token
```

```python
# Tokenization 1

tokenized_dataset = dataset_dict.map(lambda examples: tokenizer(examples["quote"], padding=True, max_length=max_length,truncation=True),batched=True)
```

```python
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)
```

```python
training_args = TrainingArguments(
    per_device_train_batch_size=16,
    per_device_eval_batch_size=8,
    num_train_epochs=10,
    learning_rate=0.0001,
    evaluation_strategy="steps",
    eval_steps=500,
    save_steps=500,
    output_dir="./quotes_model",
    overwrite_output_dir=True,
    save_total_limit=2
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["val"],
    data_collator=data_collator,
)

trainer.train()
```

```python
tokenizer.save_pretrained("quotes_model")
```

```python
model.save_pretrained("quotes_model")
```

```python
results = trainer.evaluate(tokenized_dataset["test"])
```

```python
results
```

```python
import re

def remove_non_alphanumeric(text):
    # Utilise une expression régulière pour supprimer les caractères non alphanumériques, mais conserve la ponctuation
    clean_text = re.sub(r'[^a-zA-Z0-9\s.,!?\'"]', '', text)
    return clean_text

# Générer du texte et nettoyer les caractères non alphanumériques
generated_quotes_cleaned = [remove_non_alphanumeric(text) for text in generated_quotes]

# Imprimer le texte généré nettoyé
for i, text in enumerate(generated_quotes_cleaned):
    print(f"Texte généré {i + 1} : {text}")
```

```python
def generate_joke(prompt, model, tokenizer, max_length=1000):
    # Initialisation du texte généré
    generated_text = ""
    
    # Générer du texte par morceaux jusqu'à ce qu'un point soit atteint ou que la limite maximale soit dépassée
    while len(generated_text) <= max_length:
        # Générer un morceau de texte
        generated_chunk = model.generate(
            tokenizer.encode(prompt + generated_text, return_tensors="pt").to("mps"),
            max_length=50,  # Longueur maximale d'un morceau
            num_return_sequences=1,  # Générer un seul texte à la fois
            no_repeat_ngram_size=2,  # Éviter les répétitions de n-grammes
            do_sample=True
        )[0]

        # Obtenez le texte du morceau généré
        generated_chunk_text = tokenizer.decode(generated_chunk, skip_special_tokens=True)

        # Ajoutez le morceau généré au texte complet
        generated_text += generated_chunk_text

        # Vérifiez si le texte généré contient un point
        if "." in generated_chunk_text:
            break
    
    # Nettoyez le texte généré pour supprimer les caractères non alphanumériques
    generated_text = re.sub(r'[^a-zA-Z0-9\s.,!?\'"]', '', generated_text)
    
    return generated_text

# Exemple d'utilisation de la fonction
joke = generate_joke(" Write joke :  ", model, tokenizer, max_length=1000)

# Afficher la blague générée
print("Blague générée :")
print(joke)
```
